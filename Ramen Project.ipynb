### Imported Utility  Modules needed

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import csv
import statistics
```

### Imported CSV Data into Python

```python
with open('data/ramen-ratings.csv', 'r', encoding= "utf-8") as datafile:
    ramen_data = csv.DictReader(datafile)
    data = []
    
    for row in ramen_data:
        data.append(row)
```

### Data for Plot
    I began by choosing which countries I was going to use as well as which data was going to be used from each counrty. I chose to use the style and stars data from Japan, USA, and, Malaysia. Empty lists were created for each of the different styles for each country. Then using a for loop the empty lists were filled with the stars data for each of the styles. In order to be able to graph the data the stars ratings were averaged. The averaged data was then grouped together and rounded for graphing purposes.
    

```python
my_countries = ['Japan', 'USA','Malaysia']

#Japan
cup = []
pack = []
bowl = []
box = []

#USA
cup2 = []
pack2 = []
bowl2 = []
box2 = []

#Malaysia
cup3 = []
pack3 = []
bowl3 = []
box3 = []


#Japan
for r in data:
    if r['Stars'] != 'Unrated':
        if r['Country'] == 'Japan':
            if r['Style'] == 'Cup':
                cup.append(float(r['Stars']))
            elif r['Style'] == 'Pack':
                pack.append(float(r['Stars']))
            elif r['Style'] == 'Bowl':   
                bowl.append(float(r['Stars']))
            elif r['Style'] == 'Box':
                box.append(float(r['Stars']))
    #USA           
        if r['Country'] == 'USA':
            if r['Style'] == 'Cup':
                cup2.append(float(r['Stars']))
            elif r['Style'] == 'Pack':
                pack2.append(float(r['Stars']))
            elif r['Style'] == 'Bowl':   
                bowl2.append(float(r['Stars']))
            elif r['Style'] == 'Box':
                box2.append(float(r['Stars']))

    #Malaysia
        if r['Country'] == 'Malaysia':
            if r['Style'] == 'Cup':
                cup3.append(float(r['Stars']))
            elif r['Style'] == 'Pack':
                pack3.append(float(r['Stars']))
            elif r['Style'] == 'Bowl':   
                bowl3.append(float(r['Stars']))
            elif r['Style'] == 'Box':
                box3.append(float(r['Stars']))
            
cup_mean = [statistics.mean(cup), 
              statistics.mean(cup2),
              statistics.mean(cup3)
              ]
pack_mean = [statistics.mean(pack), 
              statistics.mean(pack2),
              statistics.mean(pack3)
               ]
bowl_mean = [statistics.mean(bowl), 
              statistics.mean(bowl2),
              statistics.mean(bowl3)
               ]
box_mean = [statistics.mean(box), 
              statistics.mean(box2),
            statistics.mean(box3)
              
    
]
cup_mean= [round(x, 1) for x in cup_mean]
pack_mean= [round(x, 1) for x in pack_mean]
bowl_mean= [round(x, 1) for x in bowl_mean]
box_mean= [round(x, 1) for x in box_mean]
```

### Settings for Plot
The best way to interpret data was to use a bar graph. Each type of style was assigned a bar for each country and labled accordingly.   

```python
labels = ['Japan', 'USA','Malaysia']

x = np.arange(len(labels))  # the label locations
width = 0.14  # the width of the bars

#Bars
fig, ax = plt.subplots(figsize=(15, 10),facecolor=(.18, .31, .31))
rects1 = ax.bar(x - width/.30,cup_mean, width, label='Cup',color='#ee5c37')
rects2 = ax.bar(x - width/.45, pack_mean, width, label='Pack',color='#fff1b5')
rects3 = ax.bar(x + width/9, bowl_mean, width, label='Bowl',color='#1b2053')
rects4= ax.bar(x - width/1, box_mean, width, label='Box', color='#4bb8a6')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Stars', color='0.9',fontsize=15)
ax.set_title('Stars for Different Styles of Ramen in Three Countries', fontsize=30, color='0.9')
ax.set_xticks(x)
ax.set_xticklabels(labels, color='0.9',fontsize=15)
ax.set_facecolor('#eafff5')
ax.legend()





#lables for bars
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 3, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)
autolabel(rects4)

fig.tight_layout()

plt.ylim([0, 5.5])


plt.show()
```
